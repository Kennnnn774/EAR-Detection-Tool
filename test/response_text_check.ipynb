```python
import pandas as pd
import requests
import json

# Function to safely encode HTML content
def encode_html_content(html_content):
    return html_content.encode('utf-8', errors='replace').decode('utf-8')

# Read the CSV file
df = pd.read_csv('vulnerable_results_100k.csv')

data_list = []

for url in df['url'][:50]:
    try:
        response = requests.get(url, timeout=60, verify=False, allow_redirects=False)
        encoded_response = encode_html_content(response.text)  
        data_list.append({'URL': url, 'Response Text': encoded_response})
    except Exception as e:
        print(f"Error fetching {url}: {str(e)}")
        data_list.append({'URL': url, 'Error': str(e)})

with open('response_texts.json', 'w', encoding='utf-8') as file:
    json.dump(data_list, file, indent=4)

print("Process completed.")
```
